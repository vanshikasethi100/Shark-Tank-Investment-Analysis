```python
import pandas as pd        
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_excel(r'C:\Users\ACER\Downloads\shark_tank.xlsx')
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Anupam Investment Amount', y='Anupam Investment Equity', color='blue')

plt.title('Anupam: Investment Amount vs Equity Taken')
plt.xlabel('Amount (Lakhs)')
plt.ylabel('Equity (%)')
plt.grid(True)
plt.show()
```

```python
# 1. Find the maximum value first
max_investment = df['Anupam Investment Amount'].max()

# 2. Filter the dataframe to find the row with that max value
max_invested_company = df[df['Anupam Investment Amount'] == max_investment]

# 3. Print the name
print("The company with the maximum investment from Anupam is:")
print(max_invested_company[['Startup Name', 'Anupam Investment Amount']])
```

```python
# 1. Find the highest equity percentage he took
max_equity = df['Anupam Investment Equity'].max()

# 2. Filter the dataframe to find the startup(s)
max_equity_startup = df[df['Anupam Investment Equity'] == max_equity]

# 3. Print the name and the equity percentage
print("Startup(s) where Anupam took the maximum equity:")
print(max_equity_startup[['Startup Name', 'Anupam Investment Equity']])
```

```python
#bar chart showing the number of pitches per industry
plt.figure(figsize=(10,6)) 
sns.countplot(data=df, x='Industry')
plt.xticks(rotation=45)
plt.title('Investment Success by Industry')
plt.show()
```

```python
# 1. Your list of 7 sharks
sharks = ['Ghazal', 'Vineeta', 'Ashneer', 'Amit', 'Aman', 'Peyush', 'Namita']

# 2. A simple loop to create 7 separate plots
for shark in sharks:
    # Create a new blank chart area for each shark
    plt.figure(figsize=(8, 4))
    
    # Draw the dots
    sns.scatterplot(
        data=df, 
        x=f'{shark} Investment Amount', 
        y=f'{shark} Investment Equity',
        color='blue'
    )
    
    # Add labels so it looks professional
    plt.title(f'{shark}: Investment Amount vs Equity Taken')
    plt.xlabel('Amount (Lakhs)')
    plt.ylabel('Equity (%)')
    plt.grid(True, alpha=0.3)
    
    # Show the chart and then clear it for the next shark
    plt.show()
```
